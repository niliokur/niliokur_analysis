```python
import geopandas as gpd # for plotting on a map to display the countries

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres')) # for collecting the world map coordinates from geopandas website

for_mapping= world.merge(data, left_on = 'name', right_on = 'Country')  #merging the map and the data set prepared above

for_mapping.info() # the output shows the number of the countries common in the world map and the data set

```

```python
merged_data = for_mapping.plot(column='Life satisfaction', cmap =    
                                'Blues', figsize=(15,9), # determining the colours and the dimension of the map
                                 scheme='quantiles', k=3, legend = # k=3 refers the number of the clusters created 
                                  True,legend_kwds={'loc': 'lower left'},
                                   missing_kwds={"color": "lightgrey","edgecolor": "white",
                                                 "label": "Missing values",},); #determining the colour of the missing values(countries)

merged_data.set_title('Clustering Analysis of Women Life Satisfaction by OECD Countries', fontdict= 
            {'fontsize':25})

merged_data.set_axis_off()


```
