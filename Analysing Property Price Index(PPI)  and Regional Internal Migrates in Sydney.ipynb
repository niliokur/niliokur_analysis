```python
# data analysis
import pandas as pd

# data visualization
import matplotlib.pyplot as plt

data=pd.read_csv(r'C:\Users\berko\Desktop\datasets for projects\Residential price index Sydney.csv')
data
```

```python
fig1 = plt.figure(figsize=(20,10)) # set canvas size, set x axis longer to have a clear x axis labels.
plt.plot(data['Date'], data['Residential_PPI'],linewidth = '5') # set linewidth to have better visualization.
fig1.autofmt_xdate() #use auto formatting x axis method to avoid overlapping labels.
plt.title('Residential Property Price Index(PPI) in Sydney', loc='center', fontsize=30)
plt.grid(color= 'gray', linestyle= '--', linewidth= 0.3) #set grid lines to make more understandable graph.
plt.show()
```

```python
import datetime as dt #format how to display the date

df = pd.read_excel(r'C:\Users\berko\Desktop\datasets for projects/Regional internal migrates_Greater Sydney.xlsx')
df['Date'] = df['Date'].dt.strftime('%b-%Y') # Date was displaying %d-%m-%Y, use strftime('%b-%Y') to format datetime64 to be shown as Month-Year similar with the dataframe above.
df
```

```python
fig2 = plt.figure(figsize=(20,10)) # set canvas size, set x axis longer to havea a clear x axis labels.
plt.plot(df['Date'],df['Net_Total'],linewidth = '5') # set linewidth to have a better visualization.
plt.gca().invert_yaxis()#y axis is reversed in order to emphasize the higher negative value refers to higher migration numbers from Sydney.
                         #fig.gca()method is used to get the current axes.

fig2.autofmt_xdate() #use auto formatting x axis method to avoid overlapping labels.
plt.title('Regional Internal Migrates in Sydney', loc='center', fontsize=30)
plt.grid(color= 'gray', linestyle= '--', linewidth= 0.3) #set grid lines to make more understandable graph.
plt.show()
```

```python
fig, ax1 = plt.subplots() 
plt.rcParams["figure.figsize"] = (25,10)  
color = '#FF0000' #X11 color code of red.
ax1.set_xlabel('Date',fontsize=20) 
ax1.set_ylabel('Residential Property Price Index(PPI) in Sydney', color = color,fontsize=18) 
ax1.plot(data['Date'], data['Residential_PPI'], color = color,linewidth = '3') 
ax1.tick_params(axis ='y', labelcolor = color) 
  

ax2 = ax1.twinx() #use Twin Axes method to create a shared x axes to be used by both of the 'data' and the 'df' dataframes.
  
color = '#0000FF' #X11 color code of blue.
ax2.set_ylabel('Regional Internal Migrates in Sydney', color = color,fontsize=20) 
ax2.plot(df['Date'], df['Net_Total'], color = color, linewidth = '3') 
ax2.tick_params(axis ='y', labelcolor = color)
plt.gca().invert_yaxis() #y axis is reversed in order to emphasize the higher negative value refers to higher migration numbers from Sydney.
                         #fig.gca()method is used to get the current axes.

plt.title('Comparing Residential Property Price Index(PPI) and Regional Internal Migrates in Sydney',fontsize=30) #add title to the figure
 
plt.grid(color= 'gray', linestyle= '--', linewidth= 0.3) #set grid lines to make more understandable graph.

fig.autofmt_xdate() #use auto formatting x axis method to avoid overlapping labels.

plt.show()
```
